### B. Information storage method

I will first define a class called 'Task' and store the information of each activity as an instance. Each activity will contain eight data fields including id, description, duration, start_time, end_time, parent_task, dependency, and status. It will be possible to identify whether a task is a parent activity or a subtask by checking the value of the field 'parent_task'.


### C. Description of scheduler

By utilizing the method 'get_tasks_ready' in the class 'SimpleScheduler', I assigned 3 different priority values for each task. Since each activity has 3 subtasks, one subtask is designed to have maximum 2 dependencies. If a subtask has 2 dependencies, the priority value is 30. If a subtask has 1 dependency, the priority value is 20. If a subtask has no dependency, the priority value is 10. For the tasks that have specific time constraints, I didn't assign any priority value but created a logic that they can be started at the exact time they are supposed to begin. Therefore, the tasks with time constraints are stored in 'time_queue', which is different from 'ordinary_queue'.

A priority queue is well-suited for scheduling my tasks because it prioritizes the elements according to specified criteria. An ordinary queue is not appropriate because it usually orders elements in a first-in-first-out manner. As described above, I assign different priority values for each task and use 'heappush' method to append it into a min-heap. The 'heappop' method pulls the first task out of the min-heap and the scheduler will be able to run the appropriate task.

```python
# Imported from preclass work of CS110 Session 10
# Modifications made are explained with detailed comments

class MinHeapq:

    # Initialize an instance of MinHeapq
    # A list is expected to be passed as an argument
    def __init__(self, lst):     
        self.heap = lst
        self.heap_size = 0
        
    def left(self, i):
        return 2 * i + 1

    def right(self, i):
        return 2 * i + 2
    
    def parent(self, i):
        return (i - 1)//2
    
    def mink(self):
        
        return self.heap[0]  
    
    def heappush(self, task):
        # define dummy_task that can be compared with the input task
        dummy_task = Task(100, 'Dummy', 0, 0, 0, 0, [], float("inf"))
        self.heap.append(dummy_task)
        self.decrease_key(self.heap_size,task)
        self.heap_size+=1
        
    def decrease_key(self, i, task): 
        if task > self.heap[i]:
            raise ValueError('new key is smaller than the current key')
        self.heap[i] = task
        while i > 0 and self.heap[self.parent(i)] > self.heap[i]:
            j = self.parent(i)
            holder = self.heap[j]
            self.heap[j] = self.heap[i]
            self.heap[i] = holder
            i = j    
                 
    def heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        heap = self.heap
        if l <= (self.heap_size-1) and heap[l] < heap[i]:
            smallest = l
        else:
            smallest = i
        if r <= (self.heap_size-1) and heap[r] < heap[smallest]:
            smallest = r 
        if smallest != i:
            heap[i], heap[smallest] = heap[smallest], heap[i]
            self.heapify(smallest)
        

    def heappop(self):
        if self.heap_size < 1:
            raise ValueError('Heap underflow: There are no keys in the priority queue ')
        mink = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.heap_size-=1
        self.heapify(0)
        return mink
```

```python
# Imported from CS110 Session 10
# Modifications made are explained with detailed comments

class Task:
    """
    - id: Task id   
    - description: Short description of the task   
    - duration: Duration in minutes   
    - start_time: Start time in minutes
    - end_time: End time in minutes
    - parent_activity: Id of parent task
    - dependencies: List of the task ids that the current task depends on
    - priority: Priority level of a task (ranging from 10 to 30)   
    - status: Current status of the task      
   
    """
    
    # Define 'less than' method to compare priority values of the tasks in the queue
    def __lt__(self, other):
        return self.priority < other.priority 
    
    # Define 'representation' method that helps users to understand the instance of this class
    def __repr__(self):
        return f"\n* {self.description}\n id: {self.id}\n \tDuration:{self.duration}\n \tStart time:{self.start_time}\n \tParent activity:{self.parent_activity}\n \tPriority: {self.priority}\n \tDepends on: {self.dependencies}\n\tStatus: {self.status}"


    # Initialize an instance of Task
    def __init__(self,task_id,description,duration,start_time,end_time,parent_activity,dependencies,priority=0,status="N"):
        self.id = task_id
        self.description = description
        self.duration = duration
        self.start_time = start_time
        self.end_time = end_time
        self.parent_activity = parent_activity
        self.dependencies = dependencies
        self.priority = priority
        self.status = status
```

```python
# Imported from CS110 Session 10
# Modifications made are explained with detailed comments

class Scheduler:
    
    NOT_STARTED ='N'
    IN_ORDINARY_QUEUE = 'O'
    IN_PARENT_QUEUE = 'P'
    IN_TIME_QUEUE = 'T'
    COMPLETED = 'C'
    
    # Initialize an instance of Scheduler
    # A series of tasks is expected to be passed as an argument
    def __init__(self, tasks):
        self.tasks = tasks
        # Create a priority queue that stores the ordinary tasks
        self.ordinary_queue = []
        # Create a queue that stores parent tasks that have 3 subtasks each
        self.parent_queue = []
        # Create a queue that stores tasks that have specific time constraints
        self.time_queue = []
        # Pass the priority queue as an instance of MinHeapq
        self.my_heap = MinHeapq(self.ordinary_queue)
        
    
    def remove_dependency(self, task_id):
        """
        Remove dependency from other tasks if a task is completed
        Input: list of tasks and task_id of the task just completed
        Output: lists of tasks with t_id removed
        """
        for t in self.tasks:
            if t.id != task_id and task_id in t.dependencies:
                t.dependencies.remove(task_id)  
     
     
    def check_unscheduled_tasks(self):
        """
        Check if there is any unscheduled task and return True or False accordingly
        Input: list of tasks 
        Output: boolean (checks the status of all tasks and returns True if at least one task has status = 'N'
        """
        for task in tasks:
            if task.status == self.NOT_STARTED:
                return True
        return False 
     
        
    def check_parent_activity(self, task):
        """
        Check if all the subtasks under a parent activity are completed and print a statement if it is the case
        Input: a single task
        """
        parent_id = task.parent_activity
        children_list = []
        for i in self.ordinary_queue:
            if i.parent_activity == parent_id:
                children_list.append(i)
        for j in self.time_queue:
            if j.parent_activity == parent_id:
                children_list.append(j)
        if not children_list:
            print(f"<👋 Completed Main Task {parent_id}>\n")
            
                
    def get_tasks_ready(self):
        """ 
        Extract every task from the input and insert into an appropriate queue
        Input: list of tasks
        Output: list of tasks that are ready to execute
        """
        for task in self.tasks:
            # If a task is not yet in queue and has no parent activity
            if task.status == self.NOT_STARTED and task.parent_activity == 0:
                task.status = self.IN_PARENT_QUEUE # Change status of the task
                self.parent_queue.append(task) # append the task into the 'parent_queue'
            # If a task is not yet in queue and has a specific time constraint
            elif task.status == self.NOT_STARTED and task.start_time != 0 and not task.dependencies: 
                task.status = self.IN_TIME_QUEUE # Change status of the task
                self.time_queue.append(task) # append the task into the 'time_queue'
            # If a task is not yet in the queue, has no time constraint and no dependencies
            elif task.status == self.NOT_STARTED and task.start_time == 0 and not task.dependencies: 
                task.status = self.IN_ORDINARY_QUEUE # Change status of the task
                task.priority = 10 # assign priority value of 10
                self.my_heap.heappush(task) # heappush the task into the 'ordinary_queue'
            # If a task is not yet in the queue, has no time constraint and one dependency
            elif task.status == self.NOT_STARTED and task.start_time == 0 and len(task.dependencies) == 1: 
                task.status = self.IN_ORDINARY_QUEUE # Change status of the task
                task.priority = 20 # assign priority value of 20
                self.my_heap.heappush(task)
            # If a task is not yet in the queue, has no time constraint and two dependencies
            elif task.status == self.NOT_STARTED and task.start_time == 0 and len(task.dependencies) == 2: 
                task.status = self.IN_ORDINARY_QUEUE # Change status of the task
                task.priority = 30 # assign priority value of 30
                self.my_heap.heappush(task)
        #print(self.ordinary_queue)
        
                
    def format_time(self, time):
        return f"{time//60}h{time%60:02d}"

    
    def run_scheduler(self, starting_time = 450):
        """
        Run the actual scheduler, initiating and completing tasks while updating the time
        Input: starting_time
        """
        # Set the current time same as the starting time
        current_time = starting_time
        # Extract every task from the input and insert into an appropriate queue
        self.get_tasks_ready()
        while self.time_queue or self.ordinary_queue or self.check_unscheduled_tasks():
            # When there is a task in time_queue
            if len(self.time_queue) > 0:
                # Take the first task from the time_queue
                task = self.time_queue[0]
                # If the current time is exactly same as the starting time of the task
                if task.start_time == current_time:
                    # Pop the first task from the time_queue and execute it
                    task = self.time_queue.pop(0)
                    print(f"⏰Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration
                    print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                    self.remove_dependency(task.id)
                    task.status = self.COMPLETED
                    self.check_parent_activity(task) # Check if all the subtasks of the parent task are executed
                # If the current time is not exactly same as the starting time of the task but within 30 min range
                elif task.start_time != current_time and task.start_time - current_time <= 30:
                    # Wait until the starting time of the task
                    current_time += task.start_time - current_time
                    # Pop the first task from the time_queue and execute it
                    task = self.time_queue.pop(0)
                    print(f"⏰Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration
                    print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                    self.remove_dependency(task.id)
                    task.status = self.COMPLETED
                    self.check_parent_activity(task) # Check if all the subtasks of the parent task are executed
                # When there is a task in ordinary_queue
                elif len(self.ordinary_queue) > 0:
                    # Pop the first task from the ordinary_queue and execute it
                    task = self.my_heap.heappop()
                    print(f"⏰Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration            
                    print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                    self.remove_dependency(task.id)
                    task.status = self.COMPLETED
                    self.check_parent_activity(task) # Check if all the subtasks of the parent task are executed
            # When there is a task in ordinary_queue
            elif len(self.ordinary_queue) > 0: 
                # Pop the first task from the ordinary_queue and execute it
                task = self.my_heap.heappop()
                print(f"⏰Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                current_time += task.duration            
                print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                self.remove_dependency(task.id)
                task.status = self.COMPLETED
                self.check_parent_activity(task) # Check if all the subtasks of the parent task are executed
        total_time = current_time - starting_time             
        print(f"🏁 Completed all planned tasks in {total_time//60}h{total_time%60:02d}min")
        
```

### A. Modification from the first version

First of all, I need to add an extra feature that distinguishes whether a task could be multitasked or not. 

The first major modification would be made in 'get_tasks_ready' method. When I push each task into the heap, I assign a certain priority value based on its number of dependencies. After the process, I will add an if statement checking whether the task is multitaskable or not. If it is, I will subtract 1 from the priority value so that they can have bigger priority and be located next to each other. 

The second major modification is on 'run_scheduler' method. If the task popped from the queue is multitaskable, it will check the next task in the queue. If the next task is also multitaskable, the function will operate both tasks simultaneously and update the current time with the longer one among the durations of both tasks. The process of removing dependency, updating status, and checking parent activity will be the same. 


### B. Constraints handling

The constraints will be handled by assigning different priority values to the tasks that are in similar categories. For example, booking a hotel and reserving a restaurant can't be executed simultaneously because they both require a phone call. It is not possible to make multiple calls simultaneously. Similarly, purchasing a train ticket and making a hair shop reservation can't be executed simultaneously because they both require launching an app. It is not possible to launch multiple apps simultaneously on the same phone. Therefore, I will create a new field called 'category' and assign a different priority value to the task if it has a specific category. This will make two tasks with a similar category apart from each other in the queue, preventing them from being executed at the same time.

```python
# Define class NewScheduler that enables multitasking of certain tasks
class NewScheduler:
    
    NOT_STARTED ='N'
    IN_ORDINARY_QUEUE = 'O'
    IN_PARENT_QUEUE = 'P'
    IN_TIME_QUEUE = 'T'
    COMPLETED = 'C'
    
    
    # Initialize an instance of Scheduler
    # A series of tasks is expected to be passed as an argument
    def __init__(self, tasks):
        self.tasks = tasks
        # Create a priority queue that stores the ordinary tasks
        self.ordinary_queue = []
        # Create a queue that stores parent tasks that have 3 subtasks each
        self.parent_queue = []
        # Create a queue that stores tasks that have specific time constraints
        self.time_queue = []
        # Pass the priority queue as an instance of MinHeapq
        self.my_heap = MinHeapq(self.ordinary_queue)
        
        
    def remove_dependency(self, task_id):
        """
        Remove dependency from other tasks if a task is completed
        Input: list of tasks and task_id of the task just completed
        Output: lists of tasks with t_id removed
        """
        for t in self.tasks:
            if t.id != task_id and task_id in t.dependencies:
                t.dependencies.remove(task_id)  
      
     
    def check_unscheduled_tasks(self):
        """
        Check if there is any unscheduled task and return True or False accordingly
        Input: list of tasks 
        Output: boolean (checks the status of all tasks and returns True if at least one task has status = 'N'
        """
        for task in tasks:
            if task.status == self.NOT_STARTED:
                return True
        return False 
    
    
    def check_parent_activity(self, task):
        """
        Check if all the subtasks under a parent activity are completed and print a statement if it is the case
        Input: a single task
        """
        parent_id = task.parent_activity
        children_list = []
        for i in self.ordinary_queue:
            if i.parent_activity == parent_id:
                children_list.append(i)
        for j in self.time_queue:
            if j.parent_activity == parent_id:
                children_list.append(j)
        if not children_list:
            print(f"<👋 Completed Main Task {parent_id}>\n")
            
            
    def get_tasks_ready(self):
        """ 
        Extract every task from the input and insert into an appropriate queue
        Input: list of tasks
        Output: list of tasks that are ready to execute
        """
        for task in self.tasks:
            # If a task is not yet in queue and has no parent activity
            if task.status == self.NOT_STARTED and task.parent_activity == 0:
                task.status = self.IN_PARENT_QUEUE # Change status of the task
                self.parent_queue.append(task) # Append the task into the 'parent_queue'
            # If a task is not yet in queue and has a specific time constraint
            elif task.status == self.NOT_STARTED and task.start_time != 0 and not task.dependencies: 
                task.status = self.IN_TIME_QUEUE # Change status of the task
                self.time_queue.append(task) # Append the task into the 'time_queue'   
            # If a task is not yet in the queue, has no time constraint and no dependencies
            elif task.status == self.NOT_STARTED and task.start_time == 0 and not task.dependencies: 
                task.status = self.IN_ORDINARY_QUEUE # Change status of the task
                task.priority = 10 # Assign priority value of 10
                if task.multitasking == 'O':
                    # Assign different priority values if the task is multutaskable and has a specific category
                    if task.category == 'app_1':
                        task.priority -= 1.5
                    elif task.category == 'app_2':
                        task.priority -= 0.5
                    elif task.category == 'call_1':
                        task.priority -= 1.4
                    elif task.category == 'call_2':
                        task.priority -= 0.6
                    # Subtract 1 from the priority value if the task is multutaskable but has no specific category
                    else:
                        task.priority -= 1
                self.my_heap.heappush(task) # heappush the task into the 'ordinary_queue'
            # If a task is not yet in the queue, has no time constraint and one dependency
            elif task.status == self.NOT_STARTED and task.start_time == 0 and len(task.dependencies) == 1: 
                task.status = self.IN_ORDINARY_QUEUE # Change status of the task
                task.priority = 20 # Assign priority value of 20
                # Assign different priority values if the task is multutaskable and has a specific category
                if task.multitasking == 'O':
                    if task.category == 'app_1':
                        task.priority -= 1.5
                    elif task.category == 'app_2':
                        task.priority -= 0.5
                    elif task.category == 'call_1':
                        task.priority -= 1.4
                    elif task.category == 'call_2':
                        task.priority -= 0.6
                    # Subtract 1 from the priority value if the task is multutaskable but has no specific category
                    else:
                        task.priority -= 1
                self.my_heap.heappush(task) # heappush the task into the 'ordinary_queue'
            # If a task is not yet in the queue, has no time constraint and two dependencies
            elif task.status == self.NOT_STARTED and task.start_time == 0 and len(task.dependencies) == 2: 
                task.status = self.IN_ORDINARY_QUEUE # Change status of the task
                task.priority = 30 # Assign priority value of 30
                # Assign different priority values if the task is multutaskable and has a specific category
                if task.multitasking == 'O':
                    if task.category == 'app_1':
                        task.priority -= 1.5
                    elif task.category == 'app_2':
                        task.priority -= 0.5
                    elif task.category == 'call_1':
                        task.priority -= 1.4
                    elif task.category == 'call_2':
                        task.priority -= 0.6
                    # Subtract 1 from the priority value if the task is multutaskable but has no specific category
                    else:
                        task.priority -= 1
                self.my_heap.heappush(task) # heappush the task into the 'ordinary_queue'
       
                  
              
    def format_time(self, time):
        return f"{time//60}h{time%60:02d}"

    
    def run_scheduler(self, starting_time = 450):
        """
        Run the scheduler, initiating and completing tasks while updating the time
        Input: starting_time
        """
        # Set the current time same as the starting time
        current_time = starting_time
        # Extract every task from the input and insert into an appropriate queue
        self.get_tasks_ready()
        while self.time_queue or self.ordinary_queue or self.check_unscheduled_tasks():
            # When there is a task in time_queue
            if len(self.time_queue) > 0:
                # Take the first task from the time_queue
                task = self.time_queue[0]
                # If the current time is exactly same as the starting time of the task
                if task.start_time == current_time:
                    # Pop the first task from the time_queue and execute it
                    task = self.time_queue.pop(0)
                    print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration
                    print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                    self.remove_dependency(task.id)
                    task.status = self.COMPLETED
                    self.check_parent_activity(task) # Check if all the subtasks of the parent task are executed
                # If the current time is not exactly same as the starting time of the task but within 30 min range
                elif task.start_time != current_time and task.start_time - current_time <= 30:
                    # Wait until the starting time of the task
                    current_time += task.start_time - current_time
                    # Pop the first task from the time_queue and execute it
                    task = self.time_queue.pop(0)
                    print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration
                    print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                    self.remove_dependency(task.id)
                    task.status = self.COMPLETED
                    self.check_parent_activity(task) # Check if all the subtasks of the parent task are executed
                # When there is a task in ordinary_queue
                elif len(self.ordinary_queue) > 0:
                    # Pop the first task from the ordinary_queue
                    task = self.my_heap.heappop()
                    # Check if the task is mutitaskable
                    if task.multitasking == 'O':
                        # Check if there is a task left in the queue
                        if len(self.ordinary_queue) > 0:
                            task2 = self.ordinary_queue[0]
                            if task2.multitasking == 'O':
                                # Pop the second task if it is multitaskable
                                task2 = self.my_heap.heappop()
                                # Execute both first and second task simultaneously
                                print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                                print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task2.id} that takes {task2.duration} mins")
                                current_time += max(task.duration, task2.duration)
                                print(f"✅ Completed Task {task.id} - '{task.description}' and Task {task2.id} - '{task2.description}' at time {self.format_time(current_time)}\n")
                                self.remove_dependency(task.id)
                                self.remove_dependency(task2.id)
                                task.status = self.COMPLETED
                                task2.status = self.COMPLETED
                                self.check_parent_activity(task)
                                self.check_parent_activity(task2)
                            # If the second task is not multitaskable, just execute the first task
                            else:
                                print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                                current_time += task.duration            
                                print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                                self.remove_dependency(task.id)
                                task.status = self.COMPLETED
                                self.check_parent_activity(task)
                        # If there is no task left in the queue, just execute the first task
                        else:
                            print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                            current_time += task.duration            
                            print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                            self.remove_dependency(task.id)
                            task.status = self.COMPLETED
                            self.check_parent_activity(task)
                    # If the first task is not multitaskable, just execute the first task
                    else:
                        print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                        current_time += task.duration            
                        print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                        self.remove_dependency(task.id)
                        task.status = self.COMPLETED
                        self.check_parent_activity(task)
            # When there is no task in time queue but there is a task in ordinary queue
            elif len(self.ordinary_queue) > 0:
                # Pop the first task from the ordinary_queue
                task = self.my_heap.heappop()
                # Check if the task is mutitaskable
                if task.multitasking == 'O':
                    # Check if there is a task left in the queue
                    if len(self.ordinary_queue) > 0:
                        # Pop the second task if it is multitaskable
                        task2 = self.ordinary_queue[0]
                        if task2.multitasking == 'O':
                            # Pop the second task if it is multitaskable
                            task2 = self.my_heap.heappop()
                            # Execute both first and second task simultaneously
                            print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                            print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task2.id} that takes {task2.duration} mins")
                            current_time += max(task.duration, task2.duration)
                            print(f"✅ Completed Task {task.id} - '{task.description}' and Task {task2.id} - '{task2.description}' at time {self.format_time(current_time)}\n")
                            self.remove_dependency(task.id)
                            self.remove_dependency(task2.id)
                            task.status = self.COMPLETED
                            task2.status = self.COMPLETED
                            self.check_parent_activity(task)
                            self.check_parent_activity(task2)
                        # If the second task is not multitaskable, just execute the first task
                        else:
                            print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                            current_time += task.duration            
                            print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                            self.remove_dependency(task.id)
                            task.status = self.COMPLETED
                            self.check_parent_activity(task)
                    # If there is no task left in the queue, just execute the first task
                    else:
                        print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                        current_time += task.duration            
                        print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                        self.remove_dependency(task.id)
                        task.status = self.COMPLETED
                        self.check_parent_activity(task)
                # If the first task is not multitaskable, just execute the first task
                else:
                    print(f"⏰New Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration            
                    print(f"✅ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n") 
                    self.remove_dependency(task.id)
                    task.status = self.COMPLETED
                    self.check_parent_activity(task)
        total_time = current_time - starting_time             
        print(f"🏁 Completed all planned tasks in {total_time//60}h{total_time%60:02d}min")
        
```
